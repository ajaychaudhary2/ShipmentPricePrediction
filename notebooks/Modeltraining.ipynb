```python
import pandas as pd

def remove_outliers(X):
    # Step 1: Identify numeric columns
    numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
    
    # Step 2: Remove outliers for each numeric column
    for col in numeric_cols:
        Q1 = X[col].quantile(0.25)
        Q3 = X[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Remove rows where the numeric column has outliers
        X = X[(X[col] >= lower_bound) & (X[col] <= upper_bound)]
    
    # Return the cleaned data with both numeric and non-numeric columns
    return X

# Example usage:
# Assuming you have a DataFrame 'data' that contains your dataset
cleaned_data = remove_outliers(data)
```

```python

from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
import  numpy as np

def evaluate_model(true,predicted):
    
    mse=mean_squared_error(true,predicted)
    mae=mean_absolute_error(true,predicted)
    rmse=np.sqrt(mean_squared_error(true,predicted))
    r2_value=r2_score(true ,predicted)
    
    return mae, rmse, r2_value
```

```python
## train multiple models

models ={
    "LinearRegression":LinearRegression(),
    "Ridge":Ridge(),
    "Lasso":Lasso(),
    "ElasticNet":ElasticNet()
}
```

```python
for i in range(len(list(models))):
    
    model = list(models.values())[i]
    model.fit(X_train,y_train)
    
    
    #make prediction 
    ypred = model.predict(X_test)
    
    
    #this is  a validation or   checking the values
    
    mae,rmse,r2_value= evaluate_model(y_test,ypred)
    print(list(models.keys())[i])  
    model_list.append(list(models.keys())[i])
  
    
    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score", r2_value*100)
    
    print('='*35)

    print('\n')
    
```
